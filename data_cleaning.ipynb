# STARTING CLEANING FOR BIKE VOLUME DATA SETS

```python
# Dependencies and Setup
import pandas as pd

# Files to Load
OTT_BV_2018_LOAD = "Resources/Ottawa_Bike_Volume_2018.csv"
OTT_BV_2019_LOAD = "Resources/Ottawa_Bike_Volume_2019.csv"
OTT_BV_2020_LOAD = "Resources/Ottawa_Bike_Volume_2020.csv"
VAC_BV_2018to2020_LOAD = "Resources/Vancouver_Bike_Volume_2018-2020.csv"

# Read files and store into Pandas data frame
OTT_BV_2018 = pd.read_csv(OTT_BV_2018_LOAD)
OTT_BV_2019 = pd.read_csv(OTT_BV_2019_LOAD)
OTT_BV_2020 = pd.read_csv(OTT_BV_2020_LOAD)
VAC_BV_2018to2020 = pd.read_csv(VAC_BV_2018to2020_LOAD)
```

# Appending all Ottawa Bike Volume Date Sets into one

```python
OTT_BV_2018to2020=OTT_BV_2018.append(OTT_BV_2019)
OTT_BV_2018to2020=OTT_BV_2018to2020.append(OTT_BV_2020)
OTT_BV_2018to2020.head()
```

```python
#explore OTT_BV_2018to2020 for cleaning

#split data column to extract Day of Week, Day, Month, and Year
#split country columns
OTT_BV_2018to2020[['Day of week','Day','Month',"Year"]] = OTT_BV_2018to2020.Date.str.split(" ",expand=True)
OTT_BV_2018to2020
```

```python
#identify which columns have NAN values and how many
OTT_BV_2018to2020.isnull().sum(axis = 0)
```

```python
#1. ALEX: internal Battery failed in August 2019.
#2. LMET: Inbound counter was damaged from March 16 to April 24. (2019)

#deleting empty and not columns
OTT_BV_2018to2020=OTT_BV_2018to2020.drop(['Unnamed: 10', 'Note:',"1. ALEX: internal Battery failed in August 2019."], axis=1)
OTT_BV_2018to2020.isnull().sum(axis = 0)
```

```python
#replace the NAN value with 0 for these columns: 1^ALEX , 2^ORPY , 3^COBY ,4^CRTZ,5^LMET, 8^SOMO , 10^OGLD , 11 OBVW
OTT_BV_2018to2020['1^ALEX'] = OTT_BV_2018to2020['1^ALEX'].fillna(0)
OTT_BV_2018to2020['2^ORPY'] = OTT_BV_2018to2020['2^ORPY'].fillna(0)
OTT_BV_2018to2020['3^COBY'] = OTT_BV_2018to2020['3^COBY'].fillna(0)
OTT_BV_2018to2020['4^CRTZ'] = OTT_BV_2018to2020['4^CRTZ'].fillna(0)
OTT_BV_2018to2020['5^LMET'] = OTT_BV_2018to2020['5^LMET'].fillna(0)
OTT_BV_2018to2020['8^SOMO'] = OTT_BV_2018to2020['8^SOMO'].fillna(0)
OTT_BV_2018to2020['10^OGLD'] = OTT_BV_2018to2020['10^OGLD'].fillna(0)
OTT_BV_2018to2020['11 OBVW'] = OTT_BV_2018to2020['11 OBVW'].fillna(0)
OTT_BV_2018to2020['12a^ADAWE'] = OTT_BV_2018to2020['12a^ADAWE'].fillna(0)
OTT_BV_2018to2020['12b^ADAWE'] = OTT_BV_2018to2020['12b^ADAWE'].fillna(0)

#check null values now
OTT_BV_2018to2020.isnull().sum(axis = 0)
```

```python
#create a summed column of all bike counters
OTT_BV_2018to2020['Total Bike Volume']=OTT_BV_2018to2020.iloc[:, 1:11].sum(axis=1)
OTT_BV_2018to2020.head()
```

# Cleaning 2018-2020 Vancouver Bike Volume Data Set

```python
#Analysing Vancouver Data Set
VAC_BV_2018to2020.head()

#Splitting date
VAC_BV_2018to2020[['Month','Day',"Year"]] = VAC_BV_2018to2020.Date.str.split(" ",expand=True)
VAC_BV_2018to2020[['Day',"Comma"]] = VAC_BV_2018to2020.Day.str.split(",",expand=True)
VAC_BV_2018to2020=VAC_BV_2018to2020.drop(["Comma"], axis=1)
VAC_BV_2018to2020
```

```python
#convert string months to numeric
VAC_BV_2018to2020[["Numeric Month"]]=VAC_BV_2018to2020["Month"]
VAC_BV_2018to2020.loc[(VAC_BV_2018to2020.Month == 'January'),'Numeric Month']='1'
VAC_BV_2018to2020.loc[(VAC_BV_2018to2020.Month == 'February'),'Numeric Month']='2'
VAC_BV_2018to2020.loc[(VAC_BV_2018to2020.Month == 'March'),'Numeric Month']='3'
VAC_BV_2018to2020.loc[(VAC_BV_2018to2020.Month == 'April'),'Numeric Month']='4'
VAC_BV_2018to2020.loc[(VAC_BV_2018to2020.Month == 'May'),'Numeric Month']='5'
VAC_BV_2018to2020.loc[(VAC_BV_2018to2020.Month == 'June'),'Numeric Month']='6'
VAC_BV_2018to2020.loc[(VAC_BV_2018to2020.Month == 'July'),'Numeric Month']='7'
VAC_BV_2018to2020.loc[(VAC_BV_2018to2020.Month == 'August'),'Numeric Month']='8'
VAC_BV_2018to2020.loc[(VAC_BV_2018to2020.Month == 'September'),'Numeric Month']='9'
VAC_BV_2018to2020.loc[(VAC_BV_2018to2020.Month == 'October'),'Numeric Month']='10'
VAC_BV_2018to2020.loc[(VAC_BV_2018to2020.Month == 'November'),'Numeric Month']='11'
VAC_BV_2018to2020.loc[(VAC_BV_2018to2020.Month == 'December'),'Numeric Month']='12'
VAC_BV_2018to2020=VAC_BV_2018to2020.rename(columns={"Month": "Month-NonNumeric", "Numeric Month": "Month"})
VAC_BV_2018to2020
```

```python
VAC_BV_2018to2020["Numeric Date"]=pd.to_datetime(VAC_BV_2018to2020[['Year', 'Month', 'Day']])
VAC_BV_2018to2020["Day of Week"] = VAC_BV_2018to2020["Numeric Date"].dt.day_name()
VAC_BV_2018to2020
```

```python
#identify which columns have NAN values and how many
VAC_BV_2018to2020.isnull().sum(axis = 0)
#no null valuse found
```

```python
#create a summed column of all bike counters
VAC_BV_2018to2020['Total Bike Volume']=VAC_BV_2018to2020.iloc[:, 1:20].sum(axis=1)
VAC_BV_2018to2020.head()
```

```python
#EXPORT CLEAN DATA SETS
VAC_BV_2018to2020.to_csv("./data_sets/VAC_BV_2018to2020.csv")
OTT_BV_2018to2020.to_csv("./data_sets/OTT_BV_2018to2020.csv")
```
